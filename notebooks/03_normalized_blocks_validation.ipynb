## Chunking 전 결정 사항

- parent boundary는 `heading1` 중심으로 시작한다.
- `header`, `footer`는 검색 노이즈 후보로 별도 flag/filter를 둔다.
- 짧은 block은 인접 paragraph와 병합한다.
- child chunk는 citation recoverability를 위해 `block_id` 목록을 보존한다.
- table block은 일반 paragraph 병합과 분리해 metadata를 유지한다.
