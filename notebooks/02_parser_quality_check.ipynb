# 02 Parser Quality Check

## 목적
청킹 전에 parser/normalized block 품질을 정량 검증한다.

이 notebook은 원문 text를 출력하지 않는다. 공개 가능한 집계 지표만 확인한다.


```python
import json
from pathlib import Path

private_report = Path('..') / 'private_data' / 'reports' / 'parser_quality_report.json'
public_sample = Path('../data_samples/parser_quality_sample.json')
report_path = private_report if private_report.exists() else public_sample

report = json.loads(report_path.read_text(encoding='utf-8'))
report_path
```

```python
report['quality_summary']
```

```python
report['text_length_stats']
```

```python
sorted(report['block_count_by_element_type'].items(), key=lambda item: item[1], reverse=True)
```

## 확인할 해석

- `short_block_count`가 높으면 단독 chunk를 피하고 병합 규칙이 필요하다.
- `duplicate_text_hash_count`가 높으면 header/footer/filtering 후보를 확인한다.
- `replacement_char_block_count`가 높으면 OCR noise flag를 chunk metadata로 넘긴다.
- `p90`, `p95`는 child chunk target size 설계의 참고값이다.
