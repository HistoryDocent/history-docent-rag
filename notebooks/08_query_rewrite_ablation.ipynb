# 08 Query Rewrite Ablation

## 목적
place-aware query rewrite가 짧고 모호한 관광/음성형 질문에 도움이 되는지 검증한다.

## 비교군
- `dense_multilingual_e5_small`
- `dense_multilingual_e5_small_place_rewrite`
- `dense_multilingual_e5_small_voice_rewrite`

## 통과 기준
invalid rewrite JSON 0, voice_followup 개선, 비대상 query type 악화 없음.

## 현재 v1 결과

전체 place rewrite는 평균 Recall@5를 올렸지만 `place_fact`, `place_story`, `route_context` 일부 top-rank를 악화시켰다. v1 후보는 `voice_followup`에만 rewrite를 적용하는 `dense_multilingual_e5_small_voice_rewrite`다.

상세 결과는 `evals/reports/query_rewrite_retrieval_comparison_report.md`에 둔다.
