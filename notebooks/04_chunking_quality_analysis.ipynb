# 04 Chunking Quality Analysis

## 목적
parent/child chunk 품질과 citation recoverability를 검증한다.

## 정량 지표
- child_chunk_count
- parent_chunk_count
- orphan_child_count
- invalid_page_range_count
- unknown_element_id_count
- citation_recoverability

## 통과 기준
orphan child 0, invalid page range 0, citation recoverability 99% 이상.
